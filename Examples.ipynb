# IMDB

```python
IMDB_TEST_PATH = './dataset/imdb_test_data.parquet'
IMDB_LR_PATH = './dataset/vectors/imdb-300-LR-keys.parquet'
KEYWORD_PATH = './data/imdb-keywords/imdb_keywords.json'
        
keyword = json.load(open(KEYWORD_PATH, 'r'))
            
print('Loading...')
imdb_lrkeys = pd.read_parquet(IMDB_LR_PATH)
imdb_test = pd.read_parquet(IMDB_TEST_PATH)

            # 3. Create object to process keyword along with its connotation (keywordBank)
keywordObj = KeywordBank(keyword=keyword)
keywordObj.assign_connotation(words_len='300', 
                              class_label=['neg', 'pos'])

MODEL_PATH = '/home/anneke/Documents/models/weights/int-imdb-3-300-19-08-27_19:08:29'

imdb_result_df = pd.read_parquet(os.path.join(MODEL_PATH, 'test.parquet'))

both_correct_indices = np.where((imdb_result_df['preds'] == imdb_result_df['label']) & 
                                (imdb_result_df['preds'] != -1)&
                               (imdb_lrkeys['y_test'] == imdb_result_df['label']))[0]
LRx_reject_indices = np.where((imdb_result_df['preds'] == -1) & 
                              (imdb_lrkeys['y_test'] != imdb_result_df['label']))[0]
both_reject = np.where( (imdb_result_df['preds'] == imdb_lrkeys['y_test']) & 
                       (imdb_result_df['preds'] == -1))[0]
```

### Both correctly classify

```python
for i in both_correct_indices[:50]:
    show_explanations(imdb_result_df['preds'].iloc[i], 
                      imdb_test['X_test'].iloc[i], 
                      imdb_result_df['explanation'].iloc[i], 
                      keywordObj,
                      true_label=imdb_result_df['label'].iloc[i])
```

### LR misclassify, ours reject

```python
for i in LRx_reject_indices[3:6]:
    show_explanations(imdb_result_df['preds'].iloc[i], 
                      imdb_test['X_test'].iloc[i], 
                      imdb_result_df['explanation'].iloc[i], 
                      keywordObj,
                      true_label=imdb_result_df['label'].iloc[i])
```

### Both_Rejected

```python
for i in both_reject[:3]:
    show_explanations(imdb_result_df['preds'].iloc[i], 
                      imdb_test['X_test'].iloc[i], 
                      imdb_result_df['explanation'].iloc[i], 
                      keywordObj,
                      true_label=imdb_result_df['label'].iloc[i])
```
